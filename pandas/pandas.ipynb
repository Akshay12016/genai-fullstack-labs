```python
import pandas as pd
```

```python
data = [
    {
        "emp_id": 101,
        "emp_name": "Akshay",
        "salary": 120000,
        "department": "GenAI",
        "skills_set": ["Python", "LLMs", "LangChain"],
        "assignedornot": True,
        "client_name": "Nokia"
    },
    {
        "emp_id": 102,
        "emp_name": "Ravi",
        "salary": 90000,
        "department": "Backend",
        "skills_set": ["Java", "Spring", "SQL"],
        "assignedornot": False,
        "client_name": None
    },
    {
        "emp_id": 103,
        "emp_name": "Sita",
        "salary": 150000,
        "department": "GenAI",
        "skills_set": ["Python", "RAG", "VectorDB"],
        "assignedornot": True,
        "client_name": "Google"
    },
    {
        "emp_id": 104,
        "emp_name": "John",
        "salary": 110000,
        "department": "Frontend",
        "skills_set": ["React", "JavaScript"],
        "assignedornot": True,
        "client_name": "Nokia"
    },
    {
        "emp_id": 105,
        "emp_name": "Meena",
        "salary": 95000,
        "department": "GenAI",
        "skills_set": ["Python", "Prompt Engineering"],
        "assignedornot": False,
        "client_name": None
    }
]

df = pd.DataFrame(data)
df
```

```python
#For each department, who has the highest salary
highest_salary_df = df.loc[df.groupby("department")["salary"].idxmax()]
highest_salary_df
```

```python
#GenAI employees already assigned to a client
genai_assigned = df[
    (df["department"] == "GenAI") &
    (df["assignedornot"] == True)
]

genai_assigned
```

```python
#Create a summarized skill set column
df["summarized_skills"] = df["skills_set"].apply(
    lambda skills: ", ".join(skills[:2])
)

df[["emp_name", "summarized_skills"]]
```

```python
#Employees having Python skill and their salaries
python_employees = df[
    df["skills_set"].apply(lambda skills: "Python" in skills)
][["emp_name", "salary"]]

python_employees
```

```python
#Total employees count
total_employees = df["emp_id"].count()
total_employees
```

```python
#Count of client-assigned employees
assigned_count = df["assignedornot"].sum()
assigned_count
```

```python
#For each customer, how many employees
client_employee_count = df.groupby("client_name")["emp_id"].count()
client_employee_count
```
